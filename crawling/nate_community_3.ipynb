```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
```

```python
data = {'글제목':[0],'내용':[0],'베댓':[0]}
```

```python
table = pd.DataFrame(data)
table
```

```python
driver = webdriver.Chrome("chromedriver83.exe")
year = ['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
month = ['01','02','03','04','05','06','07','08','09','10','11','12']

for yr in year :
    for mon in month :
        driver.get('https://pann.nate.com/talk/c20002#M|1|'+yr+mon)
        time.sleep(2)
        for j in range(1,11) :
            try: 
                driver.find_elements_by_xpath("//*[@id='ppl_area']/ul/li["+str(j)+"]/a")[0].click()
                time.sleep(2)
                html = driver.page_source
                soup = BeautifulSoup(html)
                time.sleep(1)
                # 제목
                title = soup.find("meta",{"property":"og:title"})
                title = title.get('content')
                # 내용
                content = soup.find("meta",{"property":"og:description"})
                content = content.get('content')
                # 베댓
                best_coments = ''
                best_coment = soup.find("div",{"class":"cmt_best"})
                for coment in [2,5,8] :
                    try:
                        best_coments+=best_coment.find_all('span')[coment].text
                    except: continue
                data = {'글제목':[title],'내용':[content],'베댓':[best_coments]}
                table = table.append(pd.DataFrame(data))
                time.sleep(1)
                driver.get('https://pann.nate.com/talk/c20002#M|1|'+yr+mon)
                time.sleep(2)
            except: continue
```

```python
table.to_csv('네이트판_20대 이야기.csv')
```

```python
# 결혼 카데고리 주간
```

```python
data = {'글제목':[0],'내용':[0],'베댓':[0]}

table_6 = pd.DataFrame(data)
table_6

driver = webdriver.Chrome("chromedriver83.exe")
year = ['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
month = ['01','02','03','04','05','06','07','08','09','10','11','12']
week = ['1','2','3']

for yr in year :
    for mon in month :
        for wk in week : 
            driver.get('https://pann.nate.com/talk/c20025#W|1|'+yr+mon+wk)
            time.sleep(2)
            for j in range(1,11) :
                try: 
                    driver.find_elements_by_xpath("//*[@id='ppl_area']/ul/li["+str(j)+"]/a")[0].click()
                    time.sleep(2)
                    html = driver.page_source
                    soup = BeautifulSoup(html)
                    time.sleep(1)
                    # 제목
                    title = soup.find("meta",{"property":"og:title"})
                    title = title.get('content')
                    # 내용
                    content = soup.find("meta",{"property":"og:description"})
                    content = content.get('content')
                    # 베댓
                    best_coments = ''
                    best_coment = soup.find("div",{"class":"cmt_best"})
                    for coment in [2,5,8] :
                        try:
                            best_coments+=best_coment.find_all('span')[coment].text
                        except: continue
                    data = {'글제목':[title],'내용':[content],'베댓':[best_coments]}
                    table_6 = table_6.append(pd.DataFrame(data))
                    time.sleep(1)
                    driver.get('https://pann.nate.com/talk/c20025#W|1|'+yr+mon+wk)
                    time.sleep(2)
                except: continue
```

```python
table_6.to_csv('네이트판_결혼_주간.csv')
```
