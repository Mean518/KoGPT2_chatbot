```python
box = pd.DataFrame({'page':[0],'no':[0],'title':[0],'Q':[0],'A':[0],'seg':[0]},columns=['page','no','title','Q','A','seg'])
driver=webdriver.Chrome("chromedriver83.exe")
for page in range(1,80) :
    driver.get('https://www.lifeline.or.kr/business/board.php?page='+str(page)+'&gubun=public&keyfield=&keyword=&keyCate=')
    time.sleep(1)
    for writing in range(1,21) : 
        try :
            rmf = driver.find_element_by_xpath('//*[@id="form"]/div[2]/a['+str(writing)+']/h3/mark')
            rmf.click()
            time.sleep(1)
            # 제목
            title = driver.find_element_by_xpath('//*[@id="wrapper"]/article/div[3]/table/thead/tr/th').text
            time.sleep(1)
            # 내용
            question = driver.find_element_by_xpath('//*[@id="wrapper"]/article/div[3]/table/tbody/tr[2]/td').text
            time.sleep(1)
            question = question.replace('\n',' ')
            question = question.replace('  ',' ')
            if len(question) >= 500 :
                question = question[:500]
                list_dot = question.rfind('.')
                time.sleep(1)
                if list_dot != -1 : 
                    question = question[:list_dot+1]
                    time.sleep(1)
            result = spell_checker.check(question)
            question = result[2]
            # 답변
            answer = driver.find_element_by_xpath('//*[@id="wrapper"]/article/div[3]/table/tbody/tr[3]/td').text
            time.sleep(1)
            answer = answer.replace('\n',' ')
            answer = answer.replace('  ',' ')
            name_index = answer.find('님.')
            list_index = answer.find('상담원')
            answer = answer[name_index+3:list_index-2]
            
            if len(answer)>=1000 :
                time.sleep(1)
                answer = answer[-1000:]
                first_dot = answer.find('.')
                answer = answer[first_dot+2:]
                time.sleep(1)
                
            dict = {'page':[page],'no':[writing],'title':[title],'Q':[question],'A':[answer],'seg':[0]}
            box = box.append(pd.DataFrame(dict))
            print(f'{page}페이지의 {writing}번째 글 수집완료.')
            driver.get('https://www.lifeline.or.kr/business/board.php?page='+str(page)+'&gubun=public&keyfield=&keyword=&keyCate=')
        except: continue
```
